```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### Step 1: Load the dataset into a Pandas DataFrame and explore the data

```python
data = pd.read_csv('jumta.csv')
data.head(3)
```

```python
data.info()
```

```python
print(data.dtypes)
```

```python
data['Revenue'] = data['Revenue'].str.replace(',', '', regex=True)
```

```python
print(data.dtypes)
```

```python
data.head()
```

```python
data['Revenue'] = pd.to_numeric(data['Revenue'], errors='coerce')
```

```python
data.head(3)
```

```python
data.dtypes
```

```python
data.describe()
```

#### Step 2: Create a line plot showing the monthly revenue over the two-year period. Customize the plot to make it publication-quality

```python
plt.figure(figsize = (12, 5))
plt.plot(data['Month'], data['Revenue'], marker='o', label = 'Revenue', color = 'orange')
plt.xlabel('Month')
plt.ylabel('Amount of Revenue')
plt.title('Monthly Revenue 2020-2021')
plt.xticks(rotation=60)
plt.grid()
plt.legend()
plt.show()
```

#### Step 3: Create a bar plot showing the total number of orders per month. Customize the plot to make it publication-quality

```python
plt.figure(figsize = (12, 5))
plt.bar(data['Month'], data['Orders'], color = 'orange', label = 'Orders')
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.title('Monthly Orders 2020-2021')
plt.grid(axis='y')
plt.xticks(rotation=50)
plt.legend()
plt.show()
```

#### Step 4: Create a scatter plot showing the relationship between the number of orders and the average order value. Customize the plot to make it publication-quality

```python
data.head()
```

```python
plt.scatter(data['Orders'], data['Average_Order_Value'], label = 'Order Quantity', color = 'orange')
plt.xlabel('Number of Orders')
plt.ylabel('Average_Order_Value')
plt.title('Number of Orders vs Average order value (2020-2021)')
plt.grid()
plt.legend()
plt.show()
```

#### Step 5 & 6: Create a subplot containing all three visualizations created in steps 2-4. Adjust the layout and size of the subplot to make it visually appealing; and Save the subplot as a high-resolution image file

```python
fig, (ax1, ax2, ax3) = plt.subplots(3,1, figsize = (10,14))
ax1.plot(data['Month'], data['Revenue'], marker='o', label = 'Revenue', color = 'orange')
ax1.set_xlabel('Month')
ax1.set_ylabel('Amount of Revenue')
ax1.set_title('Monthly Revenue 2020-2021')
ax1.tick_params(axis='x', rotation=60)
ax1.grid()
ax1.legend()

ax2.bar(data['Month'], data['Orders'], color = 'orange', label = 'Orders')
ax2.set_xlabel('Month')
ax2.set_ylabel('Number of Orders')
ax2.set_title('Monthly Orders 2020-2021')
ax2.grid(axis='y')
ax2.tick_params(axis='x', rotation=50)
ax2.legend()

ax3.scatter(data['Orders'], data['Average_Order_Value'], label = 'Order Quantity', color = 'orange')
ax3.set_xlabel('Number of Orders')
ax3.set_ylabel('Average_Order_Value')
ax3.set_title('Number of Orders vs Average order value (2020-2021)')
ax3.grid()
ax3.legend()

plt.tight_layout()
plt.savefig('sales_data_analysis.png', dpi=300)
plt.show()
```
